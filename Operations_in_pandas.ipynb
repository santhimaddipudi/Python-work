###Label, integer, and mixed indexing

**The .loc operator**: It allows label-oriented indexing

 **The .iloc operator**: It allows integer-based indexing

 **The .ix operator**: It allows mixed label and integer-based indexing

```python
NYC_SnowAvgsData={'Months' :
['January','February','March',
'April', 'November', 'December'],
'Avg SnowDays' : [4.0,2.7,1.7,0.2,0.2,2.3],
'Avg Precip. (cm)' : [17.8,22.4,9.1,1.5,0.8,12.2],
'Avg Low Temp. (F)' : [27,29,35,45,42,32] }
NYC_SnowAvgsData
```

```python
NYC_SnowAvgs=pd.DataFrame(NYC_SnowAvgsData,
index=NYC_SnowAvgsData['Months'],
columns=['Avg SnowDays','Avg Precip. (cm)',
'Avg Low Temp. (F)'])
NYC_SnowAvgs
```

```python
#Label-oriented indexing .loc
NYC_SnowAvgs.loc['January'] #single lable
```

```python
NYC_SnowAvgs.loc[['January','April']] #multiple labels
```

```python
NYC_SnowAvgs.loc['January':'March'] #label range
```

##Note that while using the .loc, .iloc, and .ix operators on a DataFrame, the rowindex must always be specified first. This is the opposite of the
##[] operator, where
##only columns can be selected directly.

```python
#NYC_SnowAvgs.loc['Avg SnowDays']#this gives a error as mentioned column index first
NYC_SnowAvgs.loc[:,'Avg SnowDays']
```

```python
NYC_SnowAvgs.loc['April','Avg SnowDays'] #same as below
NYC_SnowAvgs.loc['March']['Avg SnowDays'] #or
NYC_SnowAvgs['Avg SnowDays']['March']#without loc and using []
```

```python
#Selection using a Boolean array
NYC_SnowAvgs.loc[NYC_SnowAvgs['Avg SnowDays']<1,:]
```

```python
SpotCrudePrices_2013.loc[:,SpotCrudePrices_2013.loc['2013-Q1']>110]
```
